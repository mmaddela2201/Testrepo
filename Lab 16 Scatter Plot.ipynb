#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Clean the data: drop rows with missing Age or JobSatPoints_6
df_clean = df.dropna(subset=['Age', 'JobSatPoints_6'])

# Convert Age to numeric if not already
df_clean['Age'] = pd.to_numeric(df_clean['Age'], errors='coerce')
df_clean = df_clean.dropna(subset=['Age'])

# Convert JobSatPoints_6 to numeric
df_clean['JobSatPoints_6'] = pd.to_numeric(df_clean['JobSatPoints_6'], errors='coerce')
df_clean = df_clean.dropna(subset=['JobSatPoints_6'])

# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df_clean['Age'], df_clean['JobSatPoints_6'], alpha=0.5, c='teal', edgecolor='k')
plt.title('Scatter Plot of Age vs. Job Satisfaction (JobSatPoints_6)', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Job Satisfaction Score (JobSatPoints_6)', fontsize=12)
plt.grid(True)
plt.show()
```

Explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) using a scatter plot.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Clean the data: Drop missing or invalid entries
df_clean = df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6'])

# Convert to numeric (in case data is read as strings)
df_clean['ConvertedCompYearly'] = pd.to_numeric(df_clean['ConvertedCompYearly'], errors='coerce')
df_clean['JobSatPoints_6'] = pd.to_numeric(df_clean['JobSatPoints_6'], errors='coerce')

# Drop rows with non-numeric values (after coercion)
df_clean = df_clean.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6'])

# Optional: Filter out extreme compensation values to avoid distortion
df_clean = df_clean[df_clean['ConvertedCompYearly'] < 500000]

# Plotting the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df_clean['ConvertedCompYearly'], df_clean['JobSatPoints_6'], alpha=0.5, color='purple', edgecolor='k')
plt.title('Scatter Plot of Yearly Compensation vs. Job Satisfaction', fontsize=14)
plt.xlabel('Yearly Compensation (USD)', fontsize=12)
plt.ylabel('Job Satisfaction Score (JobSatPoints_6)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 2: Enhancing Scatter Plots


#### 1. Scatter Plot with Trend Line for Age vs. Job Satisfaction



Add a regression line to the scatter plot of Age vs. JobSatPoints_6 to highlight trends in the data.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Map Age ranges to numeric midpoints
age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 70,
    'Prefer not to say': np.nan
}

df['AgeNum'] = df['Age'].map(age_map)

# Filter data: drop rows with NaN in AgeNum or JobSatPoints_6
df_clean = df.dropna(subset=['AgeNum', 'JobSatPoints_6'])

# Convert JobSatPoints_6 to numeric (just in case)
df_clean['JobSatPoints_6'] = pd.to_numeric(df_clean['JobSatPoints_6'], errors='coerce')

# Scatter plot
plt.figure(figsize=(10,6))
plt.scatter(df_clean['AgeNum'], df_clean['JobSatPoints_6'], alpha=0.4, label='Data points')

# Fit regression line using numpy
z = np.polyfit(df_clean['AgeNum'], df_clean['JobSatPoints_6'], 1)
p = np.poly1d(z)
plt.plot(df_clean['AgeNum'], p(df_clean['AgeNum']), color='red', linewidth=2, label='Trend Line')

plt.title('Scatter Plot with Trend Line: Age vs Job Satisfaction')
plt.xlabel('Age (years, approximate)')
plt.ylabel('Job Satisfaction (JobSatPoints_6)')
plt.legend()
plt.grid(True)
plt.show()
```

#### 2. Scatter Plot for Age vs. Work Experience


Visualize the relationship between Age (`Age`) and Work Experience (`YearsCodePro`) using a scatter plot.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Map Age ranges to numeric midpoints
age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 70,
    'Prefer not to say': np.nan
}
df['AgeNum'] = df['Age'].map(age_map)

# Convert YearsCodePro to numeric (it may contain 'Less than 1 year' or 'More than 50 years')
def convert_years_code_pro(x):
    if pd.isnull(x):
        return np.nan
    if isinstance(x, str):
        x = x.strip()
        if x == 'Less than 1 year':
            return 0.5
        elif x == 'More than 50 years':
            return 51
        else:
            try:
                return float(x)
            except:
                return np.nan
    return x

df['YearsCodeProNum'] = df['YearsCodePro'].apply(convert_years_code_pro)

# Filter to keep rows without NaN in AgeNum and YearsCodeProNum
df_clean = df.dropna(subset=['AgeNum', 'YearsCodeProNum'])

# Plot scatter plot
plt.figure(figsize=(10,6))
plt.scatter(df_clean['AgeNum'], df_clean['YearsCodeProNum'], alpha=0.5, edgecolor='k')
plt.title('Scatter Plot: Age vs Professional Coding Experience')
plt.xlabel('Age (years, approx)')
plt.ylabel('Years of Professional Coding Experience')
plt.grid(True)
plt.show()
```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Map Age categories to numeric midpoints for sizing bubbles
age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 70,
    'Prefer not to say': np.nan
}
df['AgeNum'] = df['Age'].map(age_map)

# Convert JobSatPoints_6 to numeric, coercing errors to NaN
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Filter out rows with missing data in required columns
df_clean = df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6', 'AgeNum'])

# Filter out zero or negative compensation values for clarity
df_clean = df_clean[df_clean['ConvertedCompYearly'] > 0]

# Normalize bubble size (AgeNum) for better visualization
size_scale = 20  # scaling factor for bubble size
bubble_sizes = (df_clean['AgeNum'] - df_clean['AgeNum'].min() + 1) * size_scale

# Plot bubble plot
plt.figure(figsize=(12,8))
scatter = plt.scatter(
    df_clean['ConvertedCompYearly'],
    df_clean['JobSatPoints_6'],
    s=bubble_sizes,
    alpha=0.5,
    c=df_clean['AgeNum'],  # optional: color bubbles by age for extra info
    cmap='viridis',
    edgecolors='w',
    linewidth=0.5
)

plt.colorbar(scatter, label='Age (years approx)')
plt.xlabel('Yearly Compensation (ConvertedCompYearly)')
plt.ylabel('Job Satisfaction (JobSatPoints_6)')
plt.title('Bubble Plot: Compensation vs. Job Satisfaction (Bubble size = Age)')
plt.grid(True)
plt.show()
```
